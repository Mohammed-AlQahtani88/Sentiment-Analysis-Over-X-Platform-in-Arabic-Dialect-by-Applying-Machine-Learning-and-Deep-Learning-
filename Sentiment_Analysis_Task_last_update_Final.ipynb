###Stopword Removal

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.corpus import stopwords
arabic_stopwords = set(stopwords.words('arabic'))
```

```python
# Normalize Tashkeel and Hamza in Arabic Stopwords
from pyarabic.araby import strip_tashkeel , normalize_hamza

normalized_stopwords = []
for stop_word in arabic_stopwords:
    new_sw = strip_tashkeel(stop_word)
    new_sw = normalize_hamza(new_sw, method="tasheel")
    normalized_stopwords.append(new_sw)
```

```python
# Stopword Removal
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in arabic_stopwords]
    return ' '.join(filtered_tokens)

arabic_tweets_df['tweet_stopword'] = arabic_tweets_df['tweet'].apply(remove_stopwords)
```

###Text Preprocessing

```python
import re

def preprocess_text(text):
    # Normalize Tashkeel and Hamza
    arabic_txt = strip_tashkeel(text)
    arabic_txt = normalize_hamza(arabic_txt, method="tasheel")

    # Remove digits
    arabic_txt_digits = re.sub(r'\d+', '', arabic_txt)

    # Remove non-arabic words
    arabic_txt_non_arabic = re.sub('[^\u0600-\u06FF\s]', ' ', arabic_txt_digits)

    # Remove punctuation
    arabic_txt_punc = re.sub('[ØŒØ›ØŸ.!"#$%&\'()*+,-/:;<=>?@[\]^_`{|}~]', ' ', arabic_txt_non_arabic)

    # Normalize Arabic text
    arabic_txt_norm = re.sub(r"[Ø¥Ø£Ø¢Ø§]", "Ø§", arabic_txt_punc)
    arabic_txt_norm = re.sub(r"[Ø¥Ø£Ø¢Ø§]", "Ø§", arabic_txt_norm)
    arabic_txt_norm = re.sub(r"[Ù‰]", "ÙŠ", arabic_txt_norm)
    arabic_txt_norm = re.sub(r"[Ø¤]", "Ùˆ", arabic_txt_norm)
    arabic_txt_norm = re.sub(r"[Ø¡Ø¦]", "Ø¡", arabic_txt_norm)

    return arabic_txt_norm
```

```python
arabic_tweets_df["tweets_cleaned"] = arabic_tweets_df['tweet_stopword'].apply(preprocess_text)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer  # to encode text to int
from tensorflow.keras.preprocessing.sequence import pad_sequences   # to do padding or truncating
from tensorflow.keras.models import Sequential     # the model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional   # the layers
from tensorflow.keras.callbacks import ModelCheckpoint   # save model
from tensorflow.keras.models import load_model   # load saved model
```

```python
X_data = df_processed['tweets_cleaned']
y_data = df_processed['sentiment_encoded']
```

```python
X_data
```

```python
y_data
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
```

```python
def get_max_length():
    review_length = []
    for review in X_train:
        review_length.append(len(review))

    return int(np.ceil(np.mean(review_length)))
```

```python
token = Tokenizer(lower=False)
token.fit_on_texts(X_train)
X_train = token.texts_to_sequences(X_train)
X_test = token.texts_to_sequences(X_test)

max_len = get_max_length()

X_train = pad_sequences(X_train, maxlen=max_len, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=max_len, padding='post', truncating='post')

total_words = len(token.word_index) + 1

print('Encoded X Train\n', X_train, '\n')
print('Encoded X Test\n', X_test, '\n')
print('Maximum review length: ', max_len)
print('Total words: ', total_words)
```

```python
total_words
```

```python
# ARCHITECTURE
EMBED_DIM = 32
LSTM_OUT = 64

model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=32))
model.build((None,max_len))
model.add(LSTM(LSTM_OUT, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print(model.summary())
```

```python
checkpoint = ModelCheckpoint(
    'models/LSTM.h5.keras',
    monitor='accuracy',
    save_best_only=True,
    verbose=1
)
```

```python
y_train.shape
```

```python
model.fit(X_train, y_train, batch_size = 128, epochs = 20, callbacks=[checkpoint])
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Predict probabilities
y_pred_probs = model.predict(X_test)

# Convert probabilities to binary labels (0 or 1)
y_pred = (y_pred_probs > 0.5).astype("int32").flatten()

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - LSTM Model")
plt.show()

# Classification Report
print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=["Negative", "Positive"]))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Parameters
vocab_size = 20000  # Adjust based on the dataset
max_length = 1035    # Maximum length of the tweets
oov_token = "<OOV>"  # Token for out-of-vocabulary words

# Tokenization
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(df_processed['tweets_cleaned'])  # Fit on the tweets
sequences = tokenizer.texts_to_sequences(df_processed['tweets_cleaned'])  # Convert text to sequences

# Padding
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')
```

```python
padded_sequences.shape
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Extract the labels
labels = np.array(df_processed['sentiment_encoded'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout

# Define parameters
vocab_size = 20000  # Replace with the actual size of your vocabulary
max_length = 1035  # This matches the sequence length of your data (X_train.shape[1])
embedding_dim = 1035  # Size of the embedding vector

# Model Definition
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Build the model
model.build(input_shape=(None, max_length))  # Explicitly define the input shape

# Compile the Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model Summary
model.summary()
```

```python
# Training
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
# Evaluation
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

# Predict probabilities
y_pred_probs = model.predict(X_test)

# Convert probabilities to binary predictions (0 or 1)
y_pred = (y_pred_probs > 0.5).astype("int32").flatten()

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])

# Plot confusion matrix with blue color map
disp.plot(cmap=plt.cm.Blues)  # Changed color map to Blue
plt.title("Confusion Matrix - CNN Model")
plt.show()

# Classification Report
print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=["Negative", "Positive"]))
```
